### Training First 20 day

```python
## all current data
df_daily0=df_daily[:20][['Cumulative']]
prt=int(len(df_daily0)/4)+1
weight_ls=[2.5]*prt+[1.5]*prt+[0.5]*prt+[0.1]*prt
weight_ls2=weight_ls[:len(df_daily0)]
weight_ls2.sort()
df_daily0['weight']=weight_ls2
```

```python
df_daily0.tail()
```

```python
# x=[40,3,0.8,0.9,0.1]
# logistic_model(x,[0,1,2])
# cost_func_logistic(x,df_daily0,with_pred=True)
```

```python
max_bound=[90000, 80, 150, 1.5, 7.5]
all_param,cost_value=pso_logistic(df0=df_daily0, max_bound=max_bound)
```

```python
# all_param,cost_value
for i in all_param:
    print(i)
print('.. :',cost_value)
```

```python
# all_param = [78830.1898, 67.7917167, 133.131721, 0.309796399, 5.66443831] # --> 8.4152883647
all_param = [48471.4817, 40.7540622, 101.711472, 0.307469945, 5.41640784] # --> 8.4040045827
```

```python
## Predict
T_start = 0.0
T_inc=1.0
# T_end = 23+4
T_end = 30*6

T_range = np.arange(T_start, T_end+T_inc, T_inc)
RES = logistic_model(all_param,T_range)
df_sird_pred=pd.DataFrame(RES,columns=['Cumulative_pred'])
```

```python
date_1 = dt.datetime.strptime("2020-03-02", "%Y-%m-%d")
all_date=[date_1 + dt.timedelta(days=int(i)) for i in range(T_end+1)]
```

```python
data_check=pd.concat([df_daily0[['Cumulative']],df_sird_pred[['Cumulative_pred']]],axis=1)
data_check.index=all_date
data_check[:28].plot(figsize=(10,8))
```

```python
data_check=pd.concat([df_daily[['Cumulative']],df_sird_pred[['Cumulative_pred']]],axis=1)
data_check.index=all_date
data_check[:28].plot(figsize=(10,8))
```
